## 1. Setup and Imports

```python
import numpy as np
from sklearn.metrics import accuracy_score, f1_score, classification_report

from sktime.datasets import load_italy_power_demand
from sktime_ext.classification.deep_learning.tspulse import TSPulseClassifier

from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
```

```python
import os
os.environ["PYTORCH_ENABLE_MPS_FALLBACK"] = "1"
```

```python
X_train, y_train = load_italy_power_demand(split="train", return_X_y=True)
X_test, y_test = load_italy_power_demand(split="test", return_X_y=True)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
```

## 4. Create and Configure Classifier

Create TSPulseClassifier with early stopping

```python
print("Creating TSPulseClassifier...")

clf = TSPulseClassifier(
    n_epochs=30,
    batch_size=16,              # smaller = more stable
    learning_rate=1e-4,         
    freeze_backbone=False,      # allow learning
    early_stopping=True,
    early_stopping_patience=5,
    random_state=42,
    verbose=True
)

print("\n✓ Classifier created")
print(f"Type: {type(clf).__name__}")
print(f"Epochs: {clf.n_epochs}")
print(f"Batch size: {clf.batch_size}")
print(f"Freeze backbone: {clf.freeze_backbone}")
print(f"Early stopping: {clf.early_stopping}")
```

## 5. Train the Model

Train using sktime's fit() method

```python


clf = TSPulseClassifier(
    n_epochs=50,
    batch_size=16,
    learning_rate=1e-4,
    freeze_backbone=False,
    early_stopping=True,
    early_stopping_patience=5,
    verbose=True,
    training_args={
        "no_cuda": True
    }
)
```

```python
print("Starting training...")
clf.fit(X_train, y_train)
print("Training finished ")
```

```python
print("Is fitted?", hasattr(clf, "trainer_"))
```

## 6. Make Predictions

Use predict() and predict_proba() methods

```python
print("Making predictions...")

# Predict class labels
y_pred = clf.predict(X_test)
print(f"✓ Predictions shape: {y_pred.shape}")

# Predict probabilities
y_proba = clf.predict_proba(X_test)
print(f"✓ Probabilities shape: {y_proba.shape}")

# Prediction distribution
print("\nPrediction Distribution:")
unique, counts = np.unique(y_pred, return_counts=True)

for cls, count in zip(unique, counts):
    print(f"  Class {cls}: {count}/{len(y_pred)} ({count/len(y_pred)*100:.1f}%)")
```

## 7. Evaluate Performance

Calculate accuracy, F1 score, and show classification report

```python
print("="*80)
print("EVALUATION RESULTS")
print("="*80)

# Ensure model is fitted
if not hasattr(clf, "trainer_"):
    print("Model is NOT fitted. Run clf.fit() first.")
else:
    print(" Model is fitted")

    # Predictions
    y_pred = clf.predict(X_test)
    y_proba = clf.predict_proba(X_test)

    print(f"\n✓ Predictions shape: {y_pred.shape}")
    print(f"✓ Probabilities shape: {y_proba.shape}")

    # Metrics
    acc = accuracy_score(y_test, y_pred)
    macro_f1 = f1_score(y_test, y_pred, average="macro")

    print("\nOverall Metrics:")
    print(f"  Accuracy: {acc:.4f} ({acc*100:.2f}%)")
    print(f"  Macro F1: {macro_f1:.4f} ({macro_f1*100:.2f}%)")

    # Classification report
    print("\nClassification Report:")
    print("-"*80)
    print(classification_report(y_test, y_pred))

    # Confusion matrix
    print("\nConfusion Matrix:")
    print("-"*80)
    cm = confusion_matrix(y_test, y_pred)
    print(cm)

    # Per-class accuracy
    print("\nPer-Class Accuracy:")
    print("-"*80)
    for i, cls in enumerate(clf.classes_):
        cls_acc = cm[i, i] / cm[i].sum()
        print(f"   Class {cls}: {cls_acc:.4f} ({cls_acc*100:.2f}%)")

    print("\n" + "="*80)
    print("MODEL INFORMATION")
    print("="*80)
    print(f"Model class:  {type(clf).__name__}")
    print(f"Label classes learned: {clf.classes_}")

    print("\n✓ Evaluation complete!")
```
