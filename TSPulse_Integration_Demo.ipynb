## 1. Setup and Imports

```python
import sys
import os
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import warnings
warnings.filterwarnings('ignore')

# Add paths
sys.path.insert(0, 'sktime/classification/deep_learning')
sys.path.insert(0, '../granite-tsfm')

print("✓ Imports successful")
```

```python
print("Loading dataset...")

# Try to load GunPoint (easy dataset)
dataset_name = None
try:
    from sktime.datasets import load_gunpoint
    X_train, y_train = load_gunpoint(split="train", return_X_y=True)
    X_test, y_test = load_gunpoint(split="test", return_X_y=True)
    dataset_name = "GunPoint"
    print(f"✓ Loaded {dataset_name}")
except Exception as e:
    print(f"! GunPoint not available: {e}")

# Try ItalyPowerDemand if GunPoint failed
if dataset_name is None:
    try:
        from sktime.datasets import load_italy_power_demand
        X_train, y_train = load_italy_power_demand(split="train", return_X_y=True)
        X_test, y_test = load_italy_power_demand(split="test", return_X_y=True)
        dataset_name = "ItalyPowerDemand"
        print(f"✓ Loaded {dataset_name}")
    except Exception as e:
        print(f"! ItalyPowerDemand not available: {e}")

if dataset_name is None:
    print("\n✗ No real datasets available!")
    print("Install with: pip install sktime[all_extras]")
else:
    print(f"\nDataset: {dataset_name}")
    print(f"  Train: {len(X_train)} samples")
    print(f"  Test: {len(X_test)} samples")
    print(f"  Classes: {np.unique(y_train)}")
    print(f"  Shape: {X_train.shape if hasattr(X_train, 'shape') else 'nested DataFrame'}")
```

## 4. Create and Configure Classifier

Create TSPulseClassifier with early stopping

```python
print("Creating TSPulseClassifier...")

clf = TSPulseClassifier(
    n_epochs=30,              # Training epochs
    batch_size=32,            # Batch size
    freeze_backbone=True,     # Only train classification head
    early_stopping=True,      # Enable early stopping
    early_stopping_patience=10,  # Patience
    random_state=42,          # Reproducibility
    verbose=True,             # Show progress
    tsfm_path='../granite-tsfm'  # Path to granite-tsfm
)

print("\n✓ Classifier created")
print(f"  Type: {type(clf).__name__}")
print(f"  Is BaseClassifier: {isinstance(clf, BaseClassifier)}")
print(f"\nConfiguration:")
print(f"  Epochs: {clf.n_epochs}")
print(f"  Batch size: {clf.batch_size}")
print(f"  Freeze backbone: {clf.freeze_backbone}")
print(f"  Early stopping: {clf.early_stopping}")
```

## 5. Train the Model

Train using sktime's fit() method

```python
print("="*80)
print("TRAINING")
print("="*80)

if dataset_name:
    # Train
    print(f"\nTraining on {dataset_name}...")
    clf.fit(X_train, y_train)
    
    print("\n✓ Training complete!")
    
    # Check fitted attributes (set by BaseClassifier)
    print(f"\nFitted Attributes (from BaseClassifier):")
    print(f"  classes_: {clf.classes_}")
    print(f"  n_classes_: {clf.n_classes_}")
    print(f"  _is_fitted: {clf._is_fitted}")
else:
    print("\n! Skipping training - no dataset loaded")
```

## 6. Make Predictions

Use predict() and predict_proba() methods

```python
if dataset_name:
    print("Making predictions...")
    
    # Predict class labels
    y_pred = clf.predict(X_test)
    print(f"✓ Predictions: {y_pred.shape}")
    
    # Predict probabilities
    y_proba = clf.predict_proba(X_test)
    print(f"✓ Probabilities: {y_proba.shape}")
    
    # Check prediction distribution
    print(f"\nPrediction Distribution:")
    unique, counts = np.unique(y_pred, return_counts=True)
    for cls, count in zip(unique, counts):
        print(f"  {cls}: {count}/{len(y_pred)} ({count/len(y_pred)*100:.1f}%)")
    
    # Handle label conversion if needed
    if isinstance(y_pred[0], (int, np.integer)) and not isinstance(
        y_test.iloc[0] if hasattr(y_test, 'iloc') else y_test[0], (int, np.integer)):
        print(f"\n! Converting integer predictions to original labels")
        y_pred = clf.classes_[y_pred]
else:
    print("\n! Skipping predictions - no dataset loaded")
```

## 7. Evaluate Performance

Calculate accuracy, F1 score, and show classification report

```python
if dataset_name:
    print("="*80)
    print("EVALUATION RESULTS")
    print("="*80)
    
    # Calculate metrics
    acc = accuracy_score(y_test, y_pred)
    macro_f1 = f1_score(y_test, y_pred, average='macro')
    
    # Using clf.score() - BaseClassifier method
    score = clf.score(X_test, y_test)
    
    print(f"\nOverall Metrics:")
    print(f"  Accuracy (sklearn):       {acc:.4f} ({acc*100:.2f}%)")
    print(f"  Accuracy (clf.score()):   {score:.4f} ({score*100:.2f}%)")
    print(f"  Macro F1:                 {macro_f1:.4f} ({macro_f1*100:.2f}%)")
    
    # Classification report
    print(f"\nClassification Report:")
    print("-"*80)
    print(classification_report(y_test, y_pred))
    
    # Confusion matrix
    print(f"Confusion Matrix:")
    print("-"*80)
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    
    # Interpretation
    print("\n" + "="*80)
    if acc >= 0.90:
        print("✓ EXCELLENT - Model performs very well!")
    elif acc >= 0.75:
        print("✓ GOOD - Model performs well")
    elif acc >= 0.60:
        print("! ACCEPTABLE - May need more epochs")
    else:
        print("✗ LOW - Consider increasing epochs or unfreezing backbone")
    print("="*80)
else:
    print("\n! Skipping evaluation - no dataset loaded")
```
